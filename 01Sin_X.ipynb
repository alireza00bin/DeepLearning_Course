## Import Essential Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Create Our X_train

```python
X = np.linspace(-2*np.pi, 2*np.pi, 1000).reshape(1000,1)
```

```python
y = np.sin(X)
```

## Plot Data

```python
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.axis('tight')
plt.show()
```

## Create Our ANN

We want to Train our model to predict Sin(x)

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential(
    [
        Dense(64, input_dim=X.shape[-1], activation="linear", name="input"),
        Dense(64, activation="sigmoid", name="layer1"),
        Dense(1, name="output")
    ],
    name="SIN_Model"
)
model.summary()
```

### Plot Model

```python
from keras import utils
utils.plot_model(model, show_shapes=True, show_layer_names=False)
```

### Compile Model

In this problem, we want to predict the sin(x) so We should choose the appropriate loss and optimizer

```python
model.compile(optimizer='adam', loss='mse', metrics=['mean_squared_error'])
history = model.fit(X, y, batch_size=32, epochs=500, verbose=2)
```
